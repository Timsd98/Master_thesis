```python
#read excel file containing queried proteomic data from 512c stage 
excel_table = pd.read_excel(uniprot_data, sheet_name = "without_deleted_entries")         #read excel file and select sheet
entries = excel_table.loc[:, "Entry Name"]                                                     #select only the entry numbers

#check entry numbers and print
counter = 0
for entry in json_loaded['results']:
    counter +=1
print(counter)

## check for entry query
#create list containing entrynames from json file    
data_uniprot = []                                       #create empty list to store entry name
for entry in json_loaded['results']:                    #iterate loop through json to find entry name
    data_uniprot.append(entry['uniProtkbId'])      #store entry name


#get list containing entry numbers to generate 
queried_entry_numbers = []
for entry_proteome in entries:
    if entry_proteome in data_uniprot:                  #condition to check if entry name from excel sheet is inside of uniprot data json_file
        queried_entry_numbers.append(entry_proteome)    #create list containing entry names present in json file and excel file (Proteins)
print(queried_entry_numbers)
```

```python
#create list containing queried protein sequences          
#this will be used to assign sequence to accession number                                              
queried_sequences_list = []                                 #create empty list for sequence 
queried_entry_numbers_ordered = []                          #create empty list for ordered entry numbers 

for uniprotid in json_loaded['results']:                                        #iterate loop to check if uniprotkbID = protein accession numbers
    if uniprotid['uniProtkbId'] in queried_entry_numbers:                       #check if uniprotkbId = proteinacession number
        queried_entry_numbers_ordered.append(uniprotid['uniProtkbId'])          #append list for accessionnumbers ordered
        queried_sequences_list.append(uniprotid['sequence']['value'])           #append list 


     #append empty list for sequences with sequence for entry 


print(queried_entry_numbers_ordered)
print(queried_sequences_list)
```

```python
#create json format storing Accession_number and sequence 
import json
#create empty list 
protein_data = []
#for loop to create json file from original json, using accession number and sequence 
for name, sequence in zip(queried_entry_numbers_ordered, queried_sequences_list):
    protein_data.append({"Accession_number": name, "Sequence": sequence})

print(protein_data)
```

```python
from Bio.Blast import NCBIWWW
import time                            
from tqdm import tqdm   #track progress
import os.path
from os import path
import signal 

xml_save_path = ("/Users/tim/Desktop/Important stuff/Universität/Master/4th Semester/Research Lab/Proteomic data/FASTA proteomic data/XML_files_from_Blast/")


#this is used to implement a timer for restart of the script since the server sometimes wont respond 
class TimeoutException(Exception):   # Custom exception class
    pass

def timeout_handler(signum, frame):   # Custom signal handler
    raise TimeoutException

# Change the behavior of SIGALRM
signal.signal(signal.SIGALRM, timeout_handler)
```

```python

def blast_search(saving_path, data):
    counter_blast = 0 
    for entry in tqdm(data):                #iterates for loop to go through created json
        signal.alarm(900)                     
        counter_blast +=1
        try:                                #try function tests block of code (blast function)
            if path.exists(saving_path + "" + entry['Accession_number']):             #condition to check if document already available, if so script continues with next sequence 
                print(counter_blast)
                continue
            else:                                                                     #else function to continue if file doesnt exist 
                results_handle = NCBIWWW.qblast("blastp", "nr", sequence = entry['Sequence'], format_type="Text",           #function to actually query the sequences
                matrix_name='BLOSUM62',
                word_size=5,
                expect=10,
                gapcosts='11 1',
                hitlist_size=30,
                entrez_query="Brachydanio rerio[Organism]",
    
                )
                with open(xml_save_path +  "" + entry['Accession_number'], "w") as save_file:                                  #function to save data and name it according to the fitting accession number
                    blast_results = results_handle.read() 
                    save_file.write(blast_results)
        except TimeoutException:                #exception handles error if longer than 15 min
            continue # continue the for loop if function A takes more than 15 min 

        else:
        # Reset the alarm
            signal.alarm(0)
        
        time.sleep(20)                              #incorporates 20 seconds pause 
    return blast_search      

blast_search(xml_save_path, protein_data)

```
